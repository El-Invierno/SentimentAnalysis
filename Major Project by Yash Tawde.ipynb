```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot - 1:-
sns.distplot(df['Liked'],color='red')
plt.show()

# Plot - 2:-
plt.title('No. of Positive and Negative Reviews')
plt.bar(['Positive','Negative'],[500,500],color=['green','red'])
plt.xlabel('Type of Review')
plt.ylabel('No. of Reviews')
plt.show()
```

```python
                                                        # Support Vector Classifier Model:-
x = df.iloc[:,0].values
y = df.iloc[:,1].values

# Apply train_test_split:-
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)

# Import the Count Vectorizer:-
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')

# Vectorize the x_test and x_split variables:-
x_train_vect = vect.fit_transform(x_train)
x_test_vect = vect.transform(x_test)

# Import the Support Vector Classifier (SVC):-
from sklearn.svm import SVC
model_1 = SVC()
model_1.fit(x_train_vect,y_train)

# Predict the ouput:-
y_pred_1 = model_1.predict(x_test_vect)
print(y_pred_1)

print()
# Actual output:-
print(y_test)

print()
# Accuracy Score in percentage:-
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_1) * 100
```

```python
                                                    # Pipeline Method Model (SVC + Count Vectorizer):-
# Importing Pipeline Algo:-
from sklearn.pipeline import make_pipeline

# Creation of a SVC and Count Vectorizer pipeline:-
model_2 = make_pipeline(CountVectorizer(),SVC())

# We can train the model without manually vectorizing the x variables:-
model_2.fit(x_train,y_train)

# Prediction of values using the testing data:-
y_pred_2 = model_2.predict(x_test)
print(y_pred_2)

print()
# Actual Values:-
print(y_test)

print()
# Accuracy Score in percentage:-
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_2) * 100
```

```python
                                                        # MultinomialNB Classifier method:-
# Importing the algo:-
from sklearn.naive_bayes import MultinomialNB

# Calling the MultinomialNB function.
model_3 = MultinomialNB()

# Training the model:-
model_3.fit(x_train_vect,y_train)

# Prediction of values using the model_3:-
y_pred_3 = model_3.predict(x_test_vect)
print(y_pred_3)

print()
# Actual values:-
print(y_test)

# Accuracy score in percentage:-
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_3) * 100
```

```python
                                              # Pipeline Method (CountVectorizer and MultinomialNB):-
# Importing the pipeline algo:-
from sklearn.pipeline import make_pipeline

# Creation of a pipeline:-
model_4 = make_pipeline(CountVectorizer(),MultinomialNB())

# Train the mode using the train data:-
model_4.fit(x_train,y_train)

# Predict the values:-
y_pred_4 = model_4.predict(x_test)
print(y_pred_4)

print()
# Actual Values:-
print(y_test)

# Accuracy Score in percentage:-
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_4) * 100
```

```python
# Comparison of accuracy scores of all 4 methods:-
# 1) Support Vector Classifier:- 72% (Lowest)(model_1)
# 2) Pipeline Method(Support Vector Classifier and Count Vectorization):- 79.2% (Highest)(model_2)
# 3) MultinomialNB Classfier:- 74.4% (model_3)
# 4) Pipeline Method(MultinomialNB Classifier and Count Vectorization):- 78.4% (model_4)
```
